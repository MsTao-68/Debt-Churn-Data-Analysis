```python
%matplotlib inline
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
df1 = pd.read_csv("E:/dasein_py/Data Analysis/招商银行/cleaned_features_v3.csv", index_col = 0, header =0)
df2 = pd.read_csv("E:/dasein_py/Data Analysis/招商银行/cleaned_test_v3.csv", index_col = 0, header =0)
lbl = pd.read_csv("E:/dasein_py/Data Analysis/招商银行/cleaned_labels_v3.csv", index_col = 0, header =0)
```

```python
df1.head()
```

```python
df2.head()
```

```python
label = lbl.values
```

```python
from xgboost.sklearn import XGBClassifier
from sklearn.preprocessing import StandardScaler

scaler = StandardScaler(copy=False)
df2_features = scaler.fit_transform(df2)
df1_features = scaler.fit_transform(df1)
```

```python
df1_features, df2_features
```

```python
from sklearn.model_selection import train_test_split #数据集训练集划分
x_train,x_test,y_train,y_test=train_test_split(df1_features,label,test_size=.2,random_state=22)
x_train
```

```python
y_train.reshape(-1)
```

```python
from pytorch_tabnet.tab_model import TabNetClassifier, TabNetRegressor
from pytorch_tabnet.pretraining import TabNetPretrainer
import torch
```

```python
# TabNetPretrainer 预训练模型
unsupervised_model = TabNetPretrainer(
    optimizer_fn=torch.optim.Adam,
    optimizer_params=dict(lr=2e-2),
    mask_type='sparsemax' # "sparsemax"
)

unsupervised_model.fit(
    X_train=x_train,
    eval_set=[df1.values],
    pretraining_ratio=0.8,
)

clf = TabNetClassifier(
    optimizer_fn=torch.optim.Adam,
    optimizer_params=dict(lr=2e-2),
    scheduler_params={"step_size":10, # how to use learning rate scheduler
                      "gamma":0.9},
    scheduler_fn=torch.optim.lr_scheduler.StepLR,
    mask_type='sparsemax' # This will be overwritten if using pretrain model
)

clf.fit(x_train, y_train.reshape(-1),
    eval_set= [(x_train, y_train.reshape(-1)), (df1.values, lbl.values.reshape(-1))],
    eval_name=['train', 'valid'],
    eval_metric=['auc'],
    from_unsupervised=unsupervised_model
)
```

```python
predict_proba = clf.predict_proba(x_test)
predict_proba[:,0]
```

```python
from sklearn.metrics import roc_curve, auc
false_positive_rate, true_positive_rate, thresholds = roc_curve(y_test.reshape(-1), predict_proba[:,1])
auc(false_positive_rate, true_positive_rate)
```

```python
predict_proba2 = clf.predict_proba(x_test)
false_positive_rate, true_positive_rate, thresholds = roc_curve(y_test.reshape(-1), predict_proba2[:,1])
auc(false_positive_rate, true_positive_rate)
```

```python
y_pred = clf.predict(x_test)
target_names = ['class 0', 'class 1', 'class 2']
print(classification_report(y_test, y_pred))
```
